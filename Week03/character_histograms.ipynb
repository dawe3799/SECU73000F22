```python
with open('data/loripsum.txt', 'r') as file:
    data = file.read().replace('\n', '')

data = data.lower()
    
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k',
            'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

dictionary = {}

for x in alphabet:
        dictionary[x] = data.count(x)

df = pd.DataFrame.from_dict(dictionary.items())
df.columns = ['A', 'B']

plt.bar(df.A, df.B, width=0.5, color='g')
plt.show()
```

```python
# encryption from https://www.devglan.com/online-tools/aes-encryption-decryption
import base64

plt.rcParams['figure.figsize'] = [16, 4]

alphabet_as_ord = list()
dictionary = {}

for x in range(256):
    alphabet_as_ord.append(x)
for x in alphabet_as_ord:
        dictionary[x] = 0

with open('data/loripsum_base64_enc.txt', 'r') as file:
    data = file.read().replace('\n', '')

binary_data = base64.b64decode(data)                 #decode from base64 to raw bytes (of AES)

for byte in binary_data:                             #for every byte in 'message', lookup
    dictionary[byte] = dictionary[byte] + 1          #value and increase by 1
        
df = pd.DataFrame.from_dict(dictionary.items())
df.columns = ['A', 'B']

plt.bar(df.A, df.B, width=0.5, color='g')
plt.show()
```
